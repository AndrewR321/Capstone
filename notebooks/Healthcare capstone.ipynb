```python
nhe = pd.read_csv("../data/NHE22.csv")
nhe.head(50)
```

#### Bring in columns needed & rename

```python
hcyr = (
        nhe
        [['Expenditure Amount(YEAR)',
          'National Health Expenditures (Amount in Billions)',
       'Health Consumption Expenditures', 'Personal Health Care',
       'Government Administration and Net Cost of Health Insurance',
       'Government Public Health Activities', 'Investment',
       'U.S. Population1 (Millions)',
       'Gross Domestic Product2  (Amount in Billions)']]
        )
```

```python
hcyr = hcyr.rename(columns = {'Expenditure Amount(YEAR)': 'year', 'National Health Expenditures (Amount in Billions)': 'nhe_amt_billions',
                               'Health Consumption Expenditures': 'health_consumption_exp', 'Personal Health Care': 'personal_care',
                              'Government Administration and Net Cost of Health Insurance': 'gov_adm_and_cost_health_ins',
                              'Government Public Health Activities': 'gov_public_health_activities', 'Investment': 'investment',
                              'U.S. Population1 (Millions)': 'us_pop_millions', 
                              'Gross Domestic Product2  (Amount in Billions)': 'gdp_billions'})
hcyr
```

```python
hcyr.dtypes
```

```python
hcyr['nhe_amt_billions'] = hcyr['nhe_amt_billions'].astype('int64')
```

#### Added Cost per person and the percent of change in Cost per person to dataframe

```python
hcyr['cost_per_person'] = round(hcyr['nhe_amt_billions']/(hcyr['us_pop_millions']/1000),2)
hcyr
```

```python
hcyr['yr_yr_%change'] = round(hcyr['cost_per_person'].pct_change()*100,2)
hcyr
```

```python
hcyr1 = (
        hcyr
        [['year',
          'nhe_amt_billions',
       'us_pop_millions', 'gdp_billions',
       'cost_per_person',
       'yr_yr_%change']]
        )
hcyr1
```

```python
ax=sns.barplot(data = hcyr, y='cost_per_person',  x='year')## palette= 'tab10')
ax.set_xticks(ax.get_xticks()[::5])
plt.xticks(rotation=45);
```

```python
sns.lineplot(y='cost_per_person', x='year', data = hcyr);
```

```python
sns.lineplot(y='yr_yr_%change', x='year', data = hcyr);
```

```python
sns.lineplot(data=hcyr.cost_per_person, color="g")
ax2 = plt.twinx()
sns.lineplot(data=hcyr['yr_yr_%change'], color="b", ax=ax2)
plt.title('Insurance per Person vs yoy Percent Change', fontsize = 15) ;
```

```python
ax1 = sns.set_style(style=None, rc=None)

fig, ax1 = plt.subplots(figsize=(12,6))
###ax.set_xticks(ax.get_xticks()[::5])
plt.xticks(rotation=45, fontsize = 8)
sns.lineplot(data = hcyr['yr_yr_%change'], marker='o', sort = False, ax=ax1)
ax2 = ax1.twinx()

sns.barplot(data = hcyr, x='year', y='cost_per_person', alpha=0.5, ax=ax2)
plt.title('Insurance Premium Cost vs yoy Percent Change', fontsize = 16) 
;
```
